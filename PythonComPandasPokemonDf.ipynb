```python
import pandas as pd
import matplotlib.pyplot as plt

pokemonsDf = pd.read_csv("Pokemon.csv")
display(pokemonsDf)
```

```python
#Pegando os 10 pokemons mais fortes do tipo de fogo
firePokemons = pokemonsDf.query("Type1 == 'Fire' ")
firePokemons.sort_values(by=['Attack'], ascending=False,inplace = True)
display(firePokemons.head(n=10))
```

```python
#Pegando os pokemons mais fortes de cada tipo

tipos = pokemonsDf['Type1'].unique()
pokemonsMaisFortes = []

for tipo in tipos:
  pokemons = pokemonsDf.query(f"Type1 == '{tipo}' ")
  pokemon = pokemons.sort_values(by=['Attack'], ascending=False).head(n=1)
  pokemonsMaisFortes.append(pokemon)

print(pokemonsMaisFortes)
```
