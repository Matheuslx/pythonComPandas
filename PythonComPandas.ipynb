```python
print(combustiveis_df.info())
```

```python
#Filtrando apenas uma coluna
nomesPosto = combustiveis_df["Revenda"]
print(nomesPosto)
```

```python
#Criando Dataframe com diversas colunas
colunas = combustiveis_df[['Revenda', 'Municipio','Produto', 'Valor de Venda']]
display(colunas)
```

```python
#Mostrando a linha de indice 1
display(colunas.loc[1])
```

```python
#EXIBINDO DA 10º A 20º LINHA

display(colunas.loc[9:19])
```

```python
#Pegando a gasolina mais cara
gas = colunas.loc[colunas['Produto'] == 'GASOLINA']
display(gas[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
#Pegando todos os dados cujo o produto seja etanol e o municipio seja bauru e ordenando pelo valor de venda
etanolEmBauru = colunas.loc[(colunas['Produto']=='ETANOL') & (colunas['Municipio'] == 'BAURU')]

etanolEmBauru.sort_values(by=['Valor de Venda'], inplace =True)
display(etanolEmBauru)
```

```python
#Calculando a media dos preços de venda
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), 
                            ['Valor de Venda']].mean())
```

```python
#Calculando a media do valor de venda por combustivel
valoresMediosPorCombustiveis= colunas[['Produto','Valor de Venda']].groupby(['Produto']).mean().round(2)
display(valoresMediosPorCombustiveis)
```

```python
#Adicionando coluna de ativo nas colunas
combustiveis_df['Ativo'] = True
display(combustiveis_df['Ativo'])
```

```python
#Exportando as medias por combustiveis
valoresMediosPorCombustiveis.to_excel('mediasPorCombustiveis.xlsx', sheet_name = 'Medias por combustiveis')
```

Segundo exemplo de manipulação de dados usando Python Pandas Veremos inserção de dados, remoção de colunas, merge, gráficos

```python
#Inserindo colunas
combustiveis_df['Nova coluna'] = 'Testando'
display(combustiveis_df)
```

```python
habitantes_df = pd.read_csv("habitantes_estimado.csv", sep=";")
display(habitantes_df)
```

```python
# Calcular postos de gasolina por habitante temos na amostragem de 
# combustiveis nov/2021
habitantes_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(habitantes_df)
```

```python
#Fazer um merge dos dois df's
novasColunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(habitantes_df, how = "inner", on=novasColunas)
display(combustiveis_df)
```

```python
#Dropando colunas que estao nulas
merge_df.dropna(axis= 0, inplace=True)
print(merge_df.info())
```

```python
#Agrupar e contar quantos postos tem na cidade..
postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
display(postos_por_municipio_df.info())
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "NumPostos"}, inplace=True)
display(postos_por_municipio_df)
```

```python
import matplotlib.pyplot as plt
plt.hist(combustiveis_df['Valor de Venda'])
plt.title('Valor médio por combustível')
plt.xlabel('Preço (R$)')
plt.ylabel('Quantidade de coletas')

plt.show()
```

```python
mediaCombustiveis = combustiveis_df['Valor de Venda'].groupby(by = combustiveis_df['Produto']).mean()
display(mediaCombustiveis)
```

```python
plt.figure(figsize=(10,5))
mediaCombustiveis.plot(
    kind = 'barh',
    xlabel= 'Tipo de combustivel',
    ylabel = 'Preço R$/Litro',
    color= "purple"
)

plt.show()
```

```python
from openpyxl import load_workbook
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Alignment, Color, Font


#Abrir o excel com o openxl
excel = "teste.xlsx"
mediaCombustiveis.to_excel(excel, "Sumario")
wb = load_workbook(excel)
#Pegar a planilha correta pelo nome
ws =wb['Sumario']

cinza = PatternFill("solid", fgColor="808080")
cordenadas = ['A1', 'B1']
for cordenada in cordenadas:
  ws[cordenada].fill = cinza

maxRow = ws.max_row
numLinha = 2
while(numLinha <= maxRow):
  cordenada = 'B' + str(numLinha)
  if ws[cordenada].value >=6:
    ws[cordenada].font = Font(bold=True, color = 'FF0000')
  numLinha = numLinha +1
wb.save(excel)
```
