```python
import pandas as pd

combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
#mostrando o data frame
display(combustiveis_df)
```

```python
#Exibindo as linhas da planilha
display(combustiveis_df.head())
```

```python
#Exibindo uma quantidade definida de linhas
display(combustiveis_df.head(3))
```

```python
#comandos shape e describe
#shape = retorna o numero de linhas e colunas
#describe
print("Shape:")
print(combustiveis_df.shape)
print('------------------')
print("Describe:")
print(combustiveis_df.describe())
```

```python
print(combustiveis_df.info())
```

```python
#Filtrando apenas uma coluna
nomesPosto = combustiveis_df["Revenda"]
print(nomesPosto)
```

```python
#Criando Dataframe com diversas colunas
colunas = combustiveis_df[['Revenda', 'Municipio','Produto', 'Valor de Venda']]
display(colunas)
```

```python
#Mostrando a linha de indice 1
display(colunas.loc[1])
```

```python
#EXIBINDO DA 10º A 20º LINHA

display(colunas.loc[9:19])
```

```python
#Pegando a gasolina mais cara
gas = colunas.loc[colunas['Produto'] == 'GASOLINA']
display(gas[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
#Pegando todos os dados cujo o produto seja etanol e o municipio seja bauru e ordenando pelo valor de venda
etanolEmBauru = colunas.loc[(colunas['Produto']=='ETANOL') & (colunas['Municipio'] == 'BAURU')]

etanolEmBauru.sort_values(by=['Valor de Venda'], inplace =True)
display(etanolEmBauru)
```

```python
#Calculando a media dos preços de venda
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), 
                            ['Valor de Venda']].mean())
```

```python
#Calculando a media do valor de venda por combustivel
valoresMediosPorCombustiveis= colunas[['Produto','Valor de Venda']].groupby(['Produto']).mean().round(2)
display(valoresMediosPorCombustiveis)
```

```python
#Adicionando coluna de ativo nas colunas
combustiveis_df['Ativo'] = True
display(combustiveis_df['Ativo'])
```

```python
#Exportando as medias por combustiveis
valoresMediosPorCombustiveis.to_excel('mediasPorCombustiveis.xlsx', sheet_name = 'Medias por combustiveis')
```
